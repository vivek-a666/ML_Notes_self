<h1>Seaborn is just a data visualization library like matplotlib</h1>

<h2>Relational / Statistical plots</h2>
<h4>Mainly used for multivariate analysis</h4>
<h4>- scatter plots <br> - line plots <br> - FacetGrids</h4>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

<h4>Tips is an inbuilt toy dataset in seaborn</h4>

```python
data = sns.load_dataset("tips")
```

```python
data
```

<h4>Let's plot a scatterplot between total_bill and tips</h4>

```python
sns.relplot(x = 'total_bill' , y = 'tip' , data = data)
```

<h4>The Hue Parameter</h4>

```python
sns.relplot(x = 'total_bill' , y = 'tip' , data = data , hue = 'sex')
```

<h4>The Style Parameter</h4>

```python
sns.relplot(x = 'total_bill' , y = 'tip' , data = data , hue = 'sex' , style = 'smoker')
```

<h4>The size parameter</h4>
Hue and style are mainly for categorical data but size is helpful for continuous / numerical data

```python
sns.relplot(x = 'total_bill' , y = 'tip' , data = data , hue = 'sex' , style = 'smoker' , size = 'size')
```

```python
sns.boxplot(data['total_bill'])
```

<h5>Using hue parameter</h5>

<h4>Barplot and Countplot</h4>
They give information about the central tendecy of data and not the distribution unlike they types above

```python
sns.catplot(x='sex',y='tip' , kind = "bar" , data = data)

# this shows the average tips by gender
```

```python
#hue parameter in this

sns.catplot(x='sex',y='tip' , kind = "bar" , data = data , hue = "smoker")
```

<h5>By default this graph calculates and shows median but we can change it by using estimator parameter</h5>

```python
sns.catplot(x='sex',y='tip' , kind = "bar" , data = data , hue = "smoker" , estimator=np.median)

#median
```

```python
sns.catplot(x='sex',y='tip' , kind = "bar" , data = data , hue = "smoker" , estimator=np.std)

#standard deviation
```

```python
sns.catplot(x='sex',y='tip' , kind = "bar" , data = data , hue = "smoker" , estimator=np.var)

#variance
```

<h5>Countplot is a special case of barplot which takes categorical data as input and count the number of members in each category.</h5>
It is like what histogram does for numerical data

```python
sns.catplot(x='sex' , kind='count',data=data)

#counting number of males and females
```

```python
sns.catplot(x='sex' , kind='count',data=data , hue = "smoker")

#counting number of males and females group by smoking
```

<h4>Heatmaps</h4>
individual values contained in a matrix are represented as colors

```python
#flights is another inbuilt toy dataset with seaborn
flights = sns.load_dataset("flights")
flights
```

```python
#create a pivot table

x = flights.pivot_table(index = 'year' , columns='month' , values='passengers' , aggfunc='sum')
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x , cbar = False)
```

```python
sns.heatmap(x , linewidths= 0.5)
```

```python
sns.heatmap(x , linewidths= 0.5 , annot=True , fmt='d')

# to show numbers as well
```

```python
#changing the color tone of heatmap

sns.heatmap(x , linewidths= 0.5 , annot=True , fmt='d' , cmap = 'viridis')

# more color tones are summer , winter , viridis and more in docs
```

<h4>Clustermaps</h4>
same as heatmap but hierarchically clustered

```python
sns.clustermap(x)
```

```python
#normalizing with z_score parameter
#turning on the annotation
sns.clustermap(x , z_score=0 , annot = True)
```

```python
#if we do not want to cluster either rows or columns
sns.clustermap(x , row_cluster=False)
```

```python
sns.clustermap(x , col_cluster=False)
```

<h4>Jointplots</h4>
It creates a visualization that shows the bivariate relationship between two variables and also the univariate distribution of each axis<br>
We can use them in following 4 different manners:-<br>
- Scatterplots<br>
- Hexbinplots<br>
- Kdeplots<br>
- Regplot

```python
sns.jointplot(x='total_bill' , y = 'tip' , data=data)
#scatterplot
```

```python
sns.jointplot(x='total_bill' , y = 'tip' , data=data , kind='hex')
#hexbinplot
```

```python
sns.jointplot(x='total_bill' , y = 'tip' , data=data , kind='kde')
#kdeplot
```

```python
sns.jointplot(x='total_bill' , y = 'tip' , data=data , kind="reg")
#regplot
```

<h4>Displots</h4>
 Used to do univariate analysis on numerical/continuous data<br>
They combine 3 things in a graph:-<br>
- Histogram<br>
- Kernel Density Enstimation (PDF)<br>
- Rugplot

```python
#titanic is another toy dataset with seaborn
titanic = sns.load_dataset("titanic")
titanic
```

```python
#fixing the age column a bit due to a lot of NaN values
titanic['age'].fillna(titanic['age'].mean() , inplace=True)
```

```python
titanic.sample(5)
```

```python
sns.displot(titanic['age'])
#just the histogram
```

```python
sns.displot(titanic['age'] , kde=True)
#kde also plotted
```

```python
sns.displot(titanic['age'],kde=True,rug=True)
#rugplot also true
```

```python
#we can change the bins of histogram
sns.displot(titanic['age'], bins =10)
```

```python
#plotting a displot only for those who survived and getting their age
sns.displot(titanic[titanic['survived'] ==1 ]['age'])
```
